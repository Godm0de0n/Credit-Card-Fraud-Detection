Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

Loading the Dataset

```python
# Create Dataframe with pandas
cc_data = pd.read_csv("/content/creditcard.csv")
```

```python
# Checking first 5 values
cc_data.head()
```

```python
# Checking Rows and Columns
cc_data.shape
```

```python
# Info about the Dataset
cc_data.info()
```

```python
# Distribution of Legit and Fraudulent Transaction
cc_data["Class"].value_counts()
```

Balancing the Dataset

```python
# Separating the Data for analysis
legit = cc_data[cc_data.Class == 0]
fraud = cc_data[cc_data.Class == 1]
```

```python
legit.shape
```

```python
fraud.shape
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
cc_data.groupby('Class').mean()
```

Undersampling

```python
# Build a sample dataset containing similar distribution of Legit Transaction and Fraudulent Transaction
legit_sample = legit.sample(n=1000)
```

Concatenating Tow DataFrames

```python
new_cc_data = pd.concat([legit_sample,fraud], axis =0)
```

```python
new_cc_data.shape
```

```python
new_cc_data['Class'].value_counts()
```

```python
new_cc_data.groupby('Class').mean()
```

Splitting the data into Features and Targets

```python
X = new_cc_data.drop(columns='Class', axis=1)
Y = new_cc_data['Class']
```

```python
print(X)
print(Y)
```

Split the Data into Training and Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, stratify=Y, random_state=69)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

```python
# Loading the Models
lr_model = LogisticRegression(solver='lbfgs', max_iter=100000)
rf_model = RandomForestClassifier()
```

```python
# Training LR model
lr_model.fit(X_train, Y_train)
```

```python
# Training RF model
rf_model.fit(X_train, Y_train)
```

Model Evaluation

```python
# Accuracy on Testning data
X_train_lr = lr_model.predict(X_train)
X_train_rf = rf_model.predict(X_train)
train_pred_acc_lr = accuracy_score(X_train_lr, Y_train)
train_pred_acc_rf = accuracy_score(X_train_rf, Y_train)
print(f"LR Model Accuracy (Training Data) : {train_pred_acc_lr}")
print(f"RF Model Accuracy (Training Data) : {train_pred_acc_rf}")
```

```python
# Accuracy on Testning data
X_test_lr = lr_model.predict(X_test)
X_test_rf = rf_model.predict(X_test)
test_pred_acc_lr = accuracy_score(X_test_lr, Y_test)
test_pred_acc_rf = accuracy_score(X_test_rf, Y_test)
print(f"LR Model Accuracy (Testing Data) : {test_pred_acc_lr}")
print(f"RF Model Accuracy (Testing Data) : {test_pred_acc_rf}")
```
